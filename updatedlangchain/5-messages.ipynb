## Messages

Messages are functamendal unit of context for models in LangChain. They represent the input and output models, carrying both the content and metadata needed to represent the state of conversation when interacting with an LLM. Messages are objects that contains:

1) Role
2) Content
3) Metadata

Langchain provides a standard messages type that works across all model provide, ensuring consistent behaviour regardless of the model being called.

```python
import os
from langchain.chat_models import init_chat_model

os.environ["GOOGLE_API_KEY"]= os.getenv("GOOGLE_API_KEY")
model = init_chat_model("groq:qwen/qwen3-32b")
```

```python
model.invoke("PLess tell me about artifical inelligence.")
```

### Text Prompts

Text Prompts are strings - ideal for starightforward generation tasks where you don't need to retain conversation history.

```python
model.invoke("What is Langhain Chain.")
```

Use text prompts when:

1) You have a single, standlone request
2) You don't need conversation history
3) You want minimal code complexity

### Message Prompts

Messages Types:

1) System message: Tell the model how to behave and provide context for interaction
2) Human message: Represents user input and interactions with the model
3) AI message Responses generated by AI model
4) Tool message: Requests the output of tool calls

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage

messages=[
    SystemMessage("You are the senior developer"),
    HumanMessage("How to print hello world in Java")
]

response = model.invoke(messages)
response.content
```

```python
system_msg=SystemMessage("You are a helful coding Assistant")   # we can also give detailed message

messages=[
    system_msg,
    HumanMessage("How do i create Restful API")
]
response = model.invoke(messages)
print(response.content)
```

#### Message Metadata

```python
human_msg=HumanMessage(
    content="Hello!",
    name="harsh",   # Optional: identify different user
    id="msg_123",   # Optional: unique identifier for tracing
)
```

```python
response=model.invoke([human_msg])
response
```

```python

ai_msg=AIMessage("T'd be happy to help you with the question")

message=[
    SystemMessage("You are helpful mathmatician"),
    HumanMessage("Can you help me"),
    ai_msg,
    HumanMessage("What is 2-2")
]

response=model.invoke(message)
print(response.content)
```

```python
response.usage_metadata
```

### Tool message

```python
from langchain.messages import HumanMessage, AIMessage, ToolMessage


# After a model makes a tool call
# (Here, we demonstrate manually creating the message for brevity)
ai_message=AIMessage(
    content=[],
    tool_calls=[{
        "name": "get_wheather",
        "args": {"location": "Jhansi"},
        "id": "call_123"
    }]
)

# Execute tool and create result message
wheather_result="Sunny 27 C"
tool_message=ToolMessage(
    content=wheather_result,
    tool_call_id="call_123" # Must match the call ID
)

# Continue conversation
messages=[
    HumanMessage("What is the wheather in Jhansi"),
    ai_message,
    tool_message
]

response=model.invoke(messages)
```
